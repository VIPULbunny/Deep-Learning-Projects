```python
import tensorflow as tf
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from tensorflow import keras 
```

```python
data = keras.datasets.mnist
```

```python
(X_train, y_train),(X_test,y_test) = data.load_data()
```

```python
#Flattening the X_train and X_test
```

```python
flat_X_train = X_train.reshape(len(X_train),28*28)
flat_X_test = X_test.reshape(len(X_test),28*28)
```

Unit (Neuron)

Activation

```python
# Build the model
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='sigmoid')
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(flat_X_train, y_train, epochs=5)
```

Feature Scaling


```python
new_X_train = X_train/255
new_X_test = X_test/255
```

```python
new_X_train[0][5]
```

```python
new_flat_X_train = new_X_train.reshape(len(new_X_train),28*28)
new_flat_X_test = new_X_test.reshape(len(new_X_test),28*28)
```

```python
# Train the model
model.fit(new_flat_X_train, y_train, epochs=5)
```

After Feature Scaling the accurecy got increment

```python
model.evaluate(new_flat_X_test,y_test)
```

```python
y_pred = model.predict(new_flat_X_test)
```

```python
y_pred[0]
```

```python
np.argmax(y_pred[0])
```

```python
y_pred_label = [np.argmax(i) for i in y_pred]
```

```python
#Confusion Matrics

cm = tf.math.confusion_matrix(labels=y_test ,predictions=y_pred_label)
```

```python
cm
```

```python
plt.figure(figsize=(7,5))
sns.heatmap(cm, annot=True,fmt ='d')
plt.xlabel('Actual Value')
plt.ylabel('Predicted Value')
```

```python
model = keras.Sequential([
    keras.layers.Dense(128, activation='sigmoid', input_shape=(784,)), #first hidden layer
    keras.layers.Dense(100, activation='sigmoid'),# Second Hidden layer
    keras.layers.Dense(100, activation='sigmoid')# Third Hidden layer
    # keras.layers.Dense(10, activation='softmax')  # Softmax for multi-class
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(new_flat_X_train,y_train,epochs=5)
```

```python
model = keras.Sequential([
    keras.layers.Dense(128, activation='sigmoid', input_shape=(784,)), #first hidden layer
    keras.layers.Dense(100, activation='sigmoid'),# Second Hidden layer
    keras.layers.Dense(100, activation='sigmoid'),# Third Hidden layer
    keras.layers.Dense(10, activation='softmax')  # Softmax for multi-class

])

# Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(new_flat_X_train,y_train,epochs=5)
```
